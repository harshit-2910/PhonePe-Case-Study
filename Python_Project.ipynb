# ***Task 3: Data Quality Checks***

# 3.1: Ensure data consistency across state and district levels

1. For each state, calculate the total number of transactions, total transaction
amount, and total registered users by summing up the values from the district
level data.

```python
# Group by state and calculate the totals for transactions, amount, and registered users
state_level_data = District_txn_users.groupby('State').agg(
    total_transactions=('Transactions', 'sum'),
    total_amount_inr=('Amount (INR)', 'sum'),
    total_registered_users=('Registered Users', 'sum')
).reset_index()

# Display the result
state_level_data.head()
```

2. Compare the results with the corresponding values at the state level to ensure
they match.

```python
state_txn_users_data = State_txn_users.groupby('State').agg(
    total_transactions=('Transactions', 'sum'),
    total_amount_inr=('Amount (INR)', 'sum'),
    total_registered_users=('Registered Users', 'sum')
).reset_index()

# Merge the district-level totals with the state-level totals to compare
comparison_df = pd.merge(state_level_data, state_txn_users_data, on='State', suffixes=('_district', '_state'))

# Display the result to compare the values
comparison_df.head()
```

3. Display any discrepancies found between the district-level and state-level data.

```python
# Compare the district-level and state-level totals by calculating the differences
comparison_df['discrepancy_transactions'] = comparison_df['total_transactions_district'] - comparison_df['total_transactions_state']

comparison_df['discrepancy_amount_inr'] = comparison_df['total_amount_inr_district'] - comparison_df['total_amount_inr_state']

comparison_df['discrepancy_registered_users'] = comparison_df['total_registered_users_district'] - comparison_df['total_registered_users_state']

# Filter the rows where discrepancies exist (i.e., where the difference is not zero)
discrepancies = comparison_df[(comparison_df['discrepancy_transactions'] != 0) |
                              (comparison_df['discrepancy_amount_inr'] != 0) |
                              (comparison_df['discrepancy_registered_users'] != 0)]

# Display the discrepancies
discrepancies[['State', 'discrepancy_transactions', 'discrepancy_amount_inr', 'discrepancy_registered_users']]
```

# 4.1: Ratio of users to population by state

1. Merge the State_Txn and Users dataset with the District Demographics dataset
to calculate the ratio of registered users to the population for each state.

```python
# Merge the State-TXN and Users dataset with the District-Demographics dataset
merged_data = pd.merge(State_txn_users, District_demographics, on='State', how='left')

# Group by state and aggregate the registered users and population
state_grouped = merged_data.groupby('State').agg(
    total_registered_users=('Registered Users', 'sum'),
    total_population=('Population', 'sum')
).reset_index()

# Calculate the ratio of total registered users to total population
state_grouped['user_to_population_ratio'] = state_grouped['total_registered_users'] / state_grouped['total_population']

# Display the results in a tabular format
ratio_results = state_grouped[['State', 'user_to_population_ratio']]

# Show the results
ratio_results.head()
```

2. Create a column chart depicting the ratio of users to population by state.

```python
plt.figure(figsize=(15, 6))

# Sort data for better visualization
sorted_ratio = ratio_results.sort_values(by='user_to_population_ratio', ascending=False)

# Create the column chart
plt.bar(sorted_ratio['State'], sorted_ratio['user_to_population_ratio'], color='steelblue')

# Add chart labels and title
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Users to Population Ratio')
plt.title('Ratio of Registered Users to Population by State')

plt.tight_layout()
plt.show()
```

# 4.4: Device brand usage ratio.

1. Merge the State_DeviceData dataset with the State_Txn and Users dataset.

```python
# Merging State-DeviceData with State-TXN and Users based on 'State'
merged_device_txn = pd.merge(State_devicedata, State_txn_users, on='State', how='inner')

# Display the merged result
merged_device_txn.head()
```
